    The E-Commerce company has the following rule for computing their own revenue: </font> 
    
####     The company charges  
####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600

#### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350

#### (iv) Otherwise, 5% on the final price (discounted price)    </font> 

## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
def compute_revenue(revenue):
    if revenue['discounted_price']>600:
        return revenue['discounted_price'] * .25
    elif revenue['discounted_price'] > 350 <= 600:
        return revenue['discounted_price'] * .15
    elif revenue['discounted_price'] >100 <= 350:
        return revenue['discounted_price'] * .10
    else:
        return revenue['discounted_price'] * 0.05
    
```

```python
ECom_Data['revenue']=ECom_Data.apply(compute_revenue,axis=1)
ECom_Data.head()
```

```python
Total_revenue=ECom_Data['revenue'].sum()
print('Total Revenue',Total_revenue)
```

#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.

To calculate the total brand revenue we first need to create a new column 'brand_revenue' using above condition.

```python
ECom_Data['brand_revenue']=ECom_Data['discounted_price']-ECom_Data['revenue']
ECom_Data.head()
```

```python
total_brand_revenue=ECom_Data['brand_revenue'].sum()

print('Total Brand Revenue:', total_brand_revenue)
```

```python
top_10_brands=ECom_Data.groupby('Brand')['brand_revenue'].sum().reset_index()

top_10_brands.sort_values(by='brand_revenue',ascending=False).head(10)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
ECom_Data['Region']=ECom_Data['Region'].str.strip()
ECom_Data['Order_Date']=pd.to_datetime(ECom_Data['Order_Date'],dayfirst=True)
ECom_Data['Order_Date']=pd.to_datetime(ECom_Data['Order_Date'])
ECom_Data['Month']=ECom_Data['Order_Date'].dt.month
ECom_Data['Year_Month'] = ECom_Data['Order_Date'].dt.strftime('%Y-%m')
ECom_Data.head()
```

```python
sns.lineplot(data=ECom_Data.sort_values(by='Year_Month'),x='Year_Month',y='revenue',hue='Region',style='Region',markers=True,errorbar=('ci', False))
plt.xticks(rotation=90)
plt.title('Company Monthly Revenue Regionwise',fontsize = 15,color='green');
plt.xlabel('Year_Month');
plt.ylabel('Company Revenue')
```

### Ans 9 b)


```python
plt.title('Monthly Revenue of Ecommerce Company for each region')
sns.lineplot(data=ECom_Data,x='Month',y='revenue',hue='Region',errorbar=('ci', False),style='Region',markers=True);
```
